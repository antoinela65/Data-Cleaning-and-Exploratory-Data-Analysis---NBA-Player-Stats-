## 1. Data Cleaning

The first part of this notebook is dedicated to data cleaning for the dataset named "NBA_Data_Players", which was sourced from a CSV file found on Kaggle. 

Overall, the data was relatively clean. A few rows with missing values were removed from the file prior to import. Additionally, a modification was made regarding the position data of players that occupied different positions. These changes will be explained in more details later in the project documentation. I also made several adjustments to enhance the formality and accuracy of the dataset, such as converting NaN values to N/A values and removing the leading zero from shooting percentage statistics. 

These modifications were implemented to ensure that the data format aligns with the standards on the Basketball Reference website. For reference, here is an example of the format used : https://www.basketball-reference.com/players/b/brownmo01.html


I began by importing the pandas library and loading the CSV file into the notebook using UTF-8 encoding.

```python
import pandas as pd

df = pd.read_csv('nba_data_players.csv', encoding='ISO-8859-1')

# Set the display option to show all columns for better visibility
pd.set_option('display.max_columns', None)

df
```

To improve visibility and reduce confusion, I decided to replace the NaN values with N/A values. I began by identfying the rows containing NaN values in the DataFrame. Then, I created a list of columns where the NaN values needed to be replaced. By converting the data types of these columns to string and looping through each column to perform the remplacements, I ensured that the changes were made correctly. Finally, I reviewed the DataFrame to confirm the successful completion of the process. 

```python
# Check the column that represents FG% 
fg_column = 'FG%'

# Filter rows where FG% is missing
missing_fg = df[df[fg_column].isna()]
missing_fg
```

```python
# List of columns to replace NaN values with 'N/A'
columns_to_replace = ['FG%', '3P%', '2P%', 'eFG%', 'FT%']

# Loop through each column and perform the replacement
for col in columns_to_replace:
    # Convert the column to string type
    df[col] = df[col].astype(str)
    # Replace string 'nan' with 'N/A'
    df[col].replace('nan', 'N/A', inplace=True)

df.head(90)
```

The first query was executed to confirm that there are no NaN values remaining in the dataset. The second query counted the number of N/A values in each column to provide a clearer understanding of the DataFrame.

```python
# Verify that no NaN values remain in the specified columns
nan_check = df[columns_to_replace].isna().sum()
nan_check
```

```python
# Count the number of 'N/A' values in each column
na_count = df[columns_to_replace].apply(lambda x: (x == 'N/A').sum())
na_count
```

I decided to remove the leading zero from the shooting percentage values (FG%, 2P%, 3P%, etc.) to ensure consistency and alignement with the format used on Basketball Reference. Here is how I implented this change effectively. 

```python
# Function to format decimal numbers
def format_decimal(value):
    try:
        # Convert the value to a float
        number = float(value)
        # Format the number to three decimal places
        formatted_number = f"{number:.3f}"
        # Remove leading zero if the number starts with "0"
        if formatted_number.startswith('0.'):
            return formatted_number[1:]
        return formatted_number
    except ValueError:
        # Return the original value if it cannot be converted to a float
        return value

# Process each column specified
for col in columns_to_replace:
    # Convert the column to string type to ensure consistent formatting
    df[col] = df[col].astype(str)
    # Apply the formatting function to each value in the column
    df[col] = df[col].apply(format_decimal)

df.head(90)
```

After identifying the players appearing in multiple rows in the dataset, I divided the DataFrame into two distinct groups: df_tot, which includes players who played for multiple teams, and df_other, which contains the remaining players. This separation was implemented to streamline the process of removing duplicate entries for the df_tot group.

```python
df_tot = df[df['Tm'] == 'TOT']
df_other = df[df['Tm'] != 'TOT']

df_tot
```

The following operations were performed to separate and eliminate duplicate entries in df_other. This process resulted in the creation of the df_other_cleaned DataFrame, which contains only unique entries for each player in the original DataFrame.

Finally, this set of operations was performed to merge the two groups and create the fully cleaned dataset for exploratory data analysis called df_combined. I then sorted the DataFrame by the players' last names and removed the 'First Name' column to eliminate any confusion. 

```python
# Combine df_tot and df_other_cleaned to create the full dataset
df_combined = pd.concat([df_tot, df_other_cleaned])
df_combined.head(75)

# This operation separates the player's name into two distinct columns for easier analysis and sorting with their last name.
df_combined[['First Name', 'Last Name']] = df_combined['Player'].str.split(n = 1, expand=True)

# Sort the DataFrame with the 'Last Name' column 
df_combined = df_combined.sort_values(by='Last Name')

# Removal of the'First Name' column since it is not needed anymore
df_combined = df_combined.drop(columns=['First Name'])

df_combined
```

### 2.1 Performance Metrics by Position

In this analysis, we will start by comparing the mean, median and standard deviation of points, assists and rebounds for each position. This will help to have a better understanding of the statistical trends of each positions in the NBA.

First of all, I decided to put the position in a order of my preference for visualization purposes. After calculating the mean, median and standard deviation of each stats for each position, I created separate graphs to visualize the results. Each graph shows how these statistics vary across different player positions. The comments written alongside the code will help you understand my thought process and my way of coding.

```python
# Define the categorical order for positions
position_order = ["PG", "SG", "SF", "PF", "C"]

# Convert the 'Pos' column to a categorical type with the specified order defined before
df_combined['Pos'] = pd.Categorical(df_combined['Pos'], categories=position_order, ordered=True)
df_combined
```

```python
# Group the data by position
grouped_df = df_combined.groupby('Pos')

# Define the stats to calculate in a list
stats = ['PTS', 'AST', 'TRB']

# Initialize a dictionary to store the results
results = {}

# Calculate the mean, median, and standard deviation for each position and stat
for position, group in grouped_df:
    results[position] = {}
    for stat in stats:
        results[position][f'{stat}_mean'] = group[stat].mean()
        results[position][f'{stat}_median'] = group[stat].median()
        results[position][f'{stat}_std'] = group[stat].std()

# Convert the results to a DataFrame with two decimals for easier viewing
df_results = pd.DataFrame(results).round(2)

# Transpose the results for better readability
df_results = df_results.T

# View the DataFrame
df_results
```

### 2.2 Impact of Player's Age on his Performance

Next, we will analyze how does a player's experience, based on his age in the dataset, affect his key performance metrics. We will try to investigate trends in performance metrics as a player gets older and analyze wheter younger or older players contribute more to their team's success.

### 2.3 Shooting Efficiency Analysis

This next part of the analysis will help to identify which players has the highest True Shooting Percentage (TS%) and Effective Field Goal Percentage (eFG%). 

```python
# Calculate True Shooting Percentage (TS%)
df_combined['TS%'] = df_combined['PTS'] / (2 * (df_combined['FG'] + 0.44 * df_combined['FTA']))

# Calculate Effective Field Goal Percentage (eFG%)
df_combined['eFG%'] = (df_combined['FG'] + 0.5 * df_combined['3P']) / df_combined['FGA']

# Display the DataFrame with new metrics
df_combined[['Player', 'TS%', 'eFG%']].head()
```

Next up, I will compare the efficiency of high-volume shooters and role players. 

In basketball analytics, the efficiency of players is often analyzed to understand how effectively they contribute to their team's success. When comparing high-volume shooters (players who take a large number of shots per game) to role players (players who take fewer shots but often fulfill specific duties like defense or playmaking), several key metrics and considerations come into play:

1. Defining High-Volume Shooters and Role Players

High-Volume Shooters: Typically, these are the primary offensive players on a team, often responsible for generating a significant portion of the team's points. They take many shots per game, which can sometimes lead to lower shooting percentages due to the difficulty of the shots they take.

Role Players: These players take fewer shots but often focus on other aspects of the game such as defense, rebounding, or facilitating the offense. Their shooting efficiency might be higher since they may take higher-quality shots within the flow of the offense.

```python
# Define thresholds for high-volume shooters and role players
high_volume_threshold = df_combined['PTS'].quantile(0.75)  # Top 25% in points
role_player_threshold = df_combined['PTS'].quantile(0.25)  # Bottom 25% in points

# Categorize players
df_combined['PlayerType'] = pd.cut(df_combined['PTS'], 
                                   bins=[0, role_player_threshold, high_volume_threshold, df_combined['PTS'].max()], 
                                   labels=['Role Player', 'Average Player', 'High-Volume Shooter'])
df_combined.head(25)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by PlayerType and calculate mean percentages
df_grouped = df_combined.groupby('PlayerType')[['FG%', '3P%', 'FT%']].mean().reset_index()

# Plot the results
plt.figure(figsize=(12, 6))
sns.barplot(x='PlayerType', y='FG%', data=df_grouped, palette='coolwarm')
plt.title('Field Goal Percentage by Player Type')
plt.xlabel('Player Type')
plt.ylabel('Field Goal Percentage')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='PlayerType', y='3P%', data=df_grouped, palette='coolwarm')
plt.title('Three-Point Percentage by Player Type')
plt.xlabel('Player Type')
plt.ylabel('Three-Point Percentage')
plt.show()

plt.figure(figsize=(12, 6))
sns.barplot(x='PlayerType', y='FT%', data=df_grouped, palette='coolwarm')
plt.title('Free Throw Percentage by Player Type')
plt.xlabel('Player Type')
plt.ylabel('Free Throw Percentage')
plt.show()
```

The observation that high-volume shooters have the highest FG% and 3P%, while role players have the lowest, might seem surprising because it's often assumed that players taking more difficult shots or a higher volume of shots might have lower percentages due to the increased difficulty. However, this result makes sense for several reasons:

<u>Selection Bias</u>: High-volume shooters are usually the most skilled and confident players, often given the green light to shoot frequently because they are efficient. They are more likely to have plays designed for them, allowing them to take higher-quality shots. Role players, on the other hand, may not have the same level of offensive skills and often take shots in less favorable situations.

<u>Shot Creation</u>: High-volume shooters can create their own shots or work within the offense to get better looks. Role players often rely on others to create opportunities for them, and their shot attempts might come late in the shot clock or in more contested situations.

<u>Role Definition</u>: High-volume shooters are often the primary scorers, so they focus on scoring efficiently. Role players might have other primary responsibilities like defense, rebounding, or facilitating, which means scoring efficiently is not their main focus.

<u>In summary</u>, the observation aligns with the idea that high-volume shooters, who are typically star players, are more efficient because they are more skilled and better at creating or finding high-quality shot opportunities. Role players might have lower shooting percentages because their shots come in more challenging situations or because scoring is not their primary role.

```python
# Calculate Points Per Minute
df_combined['PPM'] = df_combined['PTS'] / df_combined['MP']

# Group by PlayerType and calculate mean PPM
df_ppm_grouped = df_combined.groupby('PlayerType')['PPM'].mean().reset_index()

# Plot PPM by Player Type
plt.figure(figsize=(12, 6))
sns.barplot(x='PlayerType', y='PPM', data=df_ppm_grouped, palette='coolwarm')
plt.title('Points Per Minute by Player Type')
plt.xlabel('Player Type')
plt.ylabel('Points Per Minute')
plt.show()
```

The Points per Minute (PPM) graph provides insights into how efficiently players score relative to the time they spend on the court.

1. High-Volume Shooters: The graph highlights that high-volume shooters, typically those who play more minutes, tend to have higher points per minute. This suggests that these players are more central to their team's offense, often tasked with taking more shots and shouldering a heavier scoring load.

2. Role Players: On the other hand, role players, who generally play fewer minutes, show lower points per minute. This may be due to their limited opportunities to shoot, as they often focus on specific roles like defense or playmaking rather than scoring.

Distribution: The graph also shows the distribution of players' scoring efficiency across various minutes played. There is a clear trend that as players' minutes increase, their points per minute tend to rise, indicating that those given more time on the court are likely those who can contribute more offensively.

In conclusion, the PPM graph supports the notion that high-volume shooters are more efficient scorers compared to role players, though it's important to consider the context of each player's role within the team.

### 2.4 Defensive Metrics

In this final part of our exploratory data analysis on the NBA players' stats dataset, we will focus on defensive metrics to uncover the league's top defenders. Defense is a critical aspect of basketball that often distinguishes good teams from great ones. By analyzing various defensive statistics, we aim to identify the players who excel at stopping their opponents and making significant defensive contributions to their teams. Firstly, we will see which players, who have played at least 25 games, have the most steals and blocks per game. The 25 games played condition was added to ensure there are no outliers in the data. 

```python
# Filter out players who have played fewer than 25 games
df_filtered = df_combined[df_combined['G'] >= 25]

# Calculate average defensive metrics by position
defensive_stats = df_filtered.groupby('Pos')[['STL', 'BLK']].mean()

# Sort to find the top performers
top_steals = df_filtered[['Player', 'Pos', 'STL']].sort_values(by='STL', ascending=False).head(10)
top_blocks = df_filtered[['Player', 'Pos', 'BLK']].sort_values(by='BLK', ascending=False).head(10)

# Display the top defensive players by position
print("Top Steals per Game:\n", top_steals)
print("\nTop Blocks per Game:\n", top_blocks)
```

Here are the best defensive performers per game. Now, let's compare the defensive performance from players across different positions.  

```python
# Step 3: Group Data by Position and Calculate Averages
position_stats = df_filtered.groupby('Pos')[['STL', 'BLK']].mean().reset_index()

# Step 4: Visualize the Results
plt.figure(figsize=(14, 6))

# Plot for Steals per Game by Position
plt.subplot(1, 2, 1)
sns.barplot(x='STL', y='Pos', data=position_stats, palette='Blues_d')
plt.title('Average Steals per Game by Position')
plt.xlabel('Steals per Game')
plt.ylabel('Position')

# Plot for Blocks per Game by Position
plt.subplot(1, 2, 2)
sns.barplot(x='BLK', y='Pos', data=position_stats, palette='Greens_d')
plt.title('Average Blocks per Game by Position')
plt.xlabel('Blocks per Game')
plt.ylabel('Position')

plt.tight_layout()
plt.show()
```

This analysis compares the defensive performance of players across different positions by evaluating their average steals and blocks per game. Only players who have played at least 25 games were considered to ensure a meaningful comparison.

**Steals per Game by Position:**
- **Guards (PG, SG):** Guards, particularly point guards (PG) and shooting guards (SG), tend to have the highest steals per game. This is because guards are often more active in the perimeter defense, pressuring ball handlers and intercepting passes.
- **Forwards (SF, PF):** Small forwards (SF) and power forwards (PF) also contribute significantly to steals, though their role in defense often involves more physicality and versatility.
- **Centers (C):** Centers typically have the lowest steals per game, as their primary defensive responsibilities involve protecting the rim rather than guarding the perimeter.

**Blocks per Game by Position:**
- **Centers (C):** Centers dominate in blocks per game, as they are the last line of defense and frequently contest shots around the basket.
- **Forwards (PF, SF):** Power forwards and small forwards also contribute to blocks, with power forwards typically being more involved in rim protection.
- **Guards (PG, SG):** Guards have the lowest blocks per game, which aligns with their role in defending the perimeter rather than protecting the rim.

This comparison highlights how defensive responsibilities and strengths vary by position, with guards excelling in generating steals and centers being the primary shot blockers.
